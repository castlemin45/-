```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def cal(x,y,operator):
    if operator == '+':
        return x + y
    elif operator == '-':
        return x - y
    elif operator == '*':
        return x*y
    elif operator == '/':
        result = print('ans :', int(x/y), 'remainder is', x%y)
        return result 
        
```

```python
cal(2,3,'+')
```

```python
cal(3,2,'/')
```

```python
cal(4,2,'/')
```
